```python
def read_json_file(paths):
    for path in paths:
        try:
            with open(path, "rb") as f:  # rb: 바이너리 모드
                return f.read().decode("utf-8")
        except IOError as e:
            print("err: {}".format(e))
    return None

# 사용
ret_data = read_json_file([
    "/test/nct11afurl.json",
    "/nct11af/paws/nct11afurl.json"
])

if ret_data is not None:
    print("파일 읽기 성공")
else:
    print("모든 경로에서 파일을 읽을 수 없습니다.")
```
